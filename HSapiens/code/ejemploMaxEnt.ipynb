```python
from casadi import *

# Symbols/expressions
x = MX.sym('x')
y = MX.sym('y')
z = MX.sym('z')
f = x**2+100*z**2
g = z+(1-x)**2-y

nlp = {}                 # NLP declaration
nlp['x']= vertcat(x,y,z) # decision vars
nlp['f'] = f             # objective
nlp['g'] = g             # constraints

# Create solver instance
F = nlpsol('F','ipopt',nlp);

# Solve the problem using a guess
F(x0=[2.5,3.0,0.75],ubg=0,lbg=0)
```

```python
# Red metabólica simple: dos metabolitos (A,B) y 4 flujos (v1,..,v4)
# v1 A
# A v2 B
# A v3 B
# B v4

# Funcion objetivo: max H
# H = -sum_i (pi * log(pi) )
# pi = vi/(sum_j vj)

# Varibles, todas positivas: 0 <= vi <= 1000 

from casadi import *

# Symbols/expressions
v1 = MX.sym('v1')
v2 = MX.sym('v2')
v3 = MX.sym('v3')
v4 = MX.sym('v4')

f = -v1/(v1+v2+v3+v4)*log(v1/(v1+v2+v3+v4)) - v2/(v1+v2+v3+v4)*log(v2/(v1+v2+v3+v4)) \
    -v3/(v1+v2+v3+v4)*log(v3/(v1+v2+v3+v4)) - v4/(v1+v2+v3+v4)*log(v4/(v1+v2+v3+v4))
    
g1 = v1 - v2 - v3 # balance de masa de A
g2 = v2 + v3 - v4 # balance de masa de B

nlp = {}                  # NLP declaration
nlp['x']= vertcat(v1,v2,v3,v4) # decision vars
nlp['f'] = -f              # objective
nlp['g'] = vertcat(g1,g2) # constraints

# Create solver instance
F = nlpsol('F','ipopt',nlp);

# Solve the problem using a guess
F(x0=[10,10,0,10],ubg=[0,0],lbg=[0,0],ubx=[10,1000,1000,1000],lbx=[10,0,0,0])
```
