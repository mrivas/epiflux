```python
import numpy as np
import pandas as pd
```

```python
#Cargar archivo gtf con anotaciones del genoma de la levadura, esto nos servira para correlacionar con cada chip-seq 
gtf= pd.read_csv("/home/bguzman/jupyter/UI/YMC/annotation/Saccer2.txt", sep="\t", lineterminator='\n', squeeze=True).to_dict()
```

```python
# segregar ID de genes
genes= gtf["gene_id\r"]
lista_id=[]
for k,v in genes.items():
    name=v.split(';')[0]
    if name not in lista_id:
        lista_id.append(name)
    
#for i in lista_id:
 #   print (i)
    
```

```python
# segregar ID de genes
genes2= lista_id
lista_id2=[]
for k in genes2:
    name2=k.split('gene_id_')[1]
    lista_id2.append(name2)
    
lista_id2

    
```

```python
#Cargar gtf como df y no como dicc
asd= pd.read_csv("/home/bguzman/jupyter/UI/YMC/annotation/Saccer2.txt", sep="\t")
```

```python
asd
```

```python
#dejar solo las fijas que contengan la informacion de los transcritos
delete_row = asd[asd.iloc[:,1]!="transcript"].index
asd = asd.drop(delete_row)
```

```python
asd
```

```python
#Crear listas para convertirlas en df
lista_chrm=[]
for i in asd["1"]:
    lista_chrm.append(i)
```

```python
lista_chrm
```

```python
#Si los datos de la columna 5(4 en python) son "+" guardar el valor del dato de la columna 2, en cambio 
#si los datos de la columna 5(4 en python) son "-" guardar el valor del dato de la columna 3.
lista_trans=[] 
for i in range(7130):
    if asd.iloc[i][4]=="+":
        lista_trans.append(asd.iloc[i][2])
    else:
        lista_trans.append(asd.iloc[i][3])
        
lista_trans
```

```python
#restar 500 para crear el rango de la cota inf.
lista_1_=[]
for i in lista_trans:
    if i >=500:
        i=i-500
        lista_1_.append(i)
    else:
        i=i-i
        lista_1_.append(i)
        
lista_1_
```

```python
#restar 500 , pero ahora redondeando al centesimo mas cercano
lista_1=[]
for i in lista_trans:
    if i >=500:
        i=i-500
        lista_1.append(round(i,-2))
    else:
        i=i-i
        lista_1.append(round(i,-2))
        
lista_1
```

```python
#sumar 500 para buscar el rango de la cota_sup
lista_2_=[]
for i in lista_trans:
        i=i+500
        lista_2_.append(i)

        
lista_2_
```

```python
#Sumar 500 y redondear al centesimo mas cercano
lista_2=[]
for i in lista_trans:
        i=i+500
        lista_2.append(round(i,-2))

        
lista_2
```

```python
for i in lista_2:
    i=round(i, -2)

print(i)
```

```python
#construccion de los rangos en los que analizaremos los chip-seq
chrm = lista_chrm 
start = lista_1
end = lista_2
gene_id= lista_id2
dict = {'chrm': chrm, 'start': start, 'end': end, 'gene_id': gene_id} 
df = pd.DataFrame(dict) 
df.to_csv('df.csv', index=False)
```
